```python
def findout_outliers(data_frame,col_name):
    global Q3,Q1,IQR_median,lower_bound,upper_bound
    Q1=data_frame[col_name].quantile(.25)
    Q3=data_frame[col_name].quantile(.75)
    IQR_median=(Q3-Q1)
    lower_bound=Q1-1.5*IQR_median
    upper_bound=Q3+1.5*IQR_median
    return ((data_frame[col_name]<lower_bound) | (data_frame[col_name]>upper_bound)).sum()
print("="*50)
print("Customer Details Dataset DataFrame:")
print("Customer_details_df with age:",findout_outliers(customer_details_df,'age'))
print("Customer_details_df with driving licence present:",findout_outliers(customer_details_df,'driving licence present'))
print("Customer_details_df with region code:",findout_outliers(customer_details_df,'region code'))
print("Customer_details_df with previously_insured:",findout_outliers(customer_details_df,'previously_insured'))
print("="*50)
print("Customer Policy  Details Dataset DataFrame:")
print("customer_policy_details_df with annual_premium:",findout_outliers(customer_policy_details_df,'annual_premium'))
print("customer_policy_details_df with sales channel code:",findout_outliers(customer_policy_details_df,'sales channel code'))
print("customer_policy_details_df with vintage:",findout_outliers(customer_policy_details_df,'vintage'))
print("customer_policy_details_df with response:",findout_outliers(customer_policy_details_df,'response'))
print("="*50)
```

```python
#-Replace all outlier values for numeric columns by mean.
def replace_outliers(data_frame,col_name):
    findout_outliers(data_frame,col_name)
    outliers= ~data_frame[col_name].between(lower_bound,upper_bound)
    mean=data_frame.loc[~outliers,col_name].mean()
    data_frame.loc[outliers,col_name]=mean
    
    return ((data_frame[col_name]<lower_bound) | (data_frame[col_name]>upper_bound)).sum()
    
print("="*50)
print("Customer Details Dataset DataFrame:")
print("Customer_details_df with age:",replace_outliers(customer_details_df,'age'))
print("Customer_details_df with driving licence present:",replace_outliers(customer_details_df,'driving licence present'))
print("Customer_details_df with region code:",replace_outliers(customer_details_df,'region code'))
print("Customer_details_df with previously_insured:",replace_outliers(customer_details_df,'previously_insured'))
print("="*50)
print("Customer Policy  Details Dataset DataFrame:")
print("customer_policy_details_df with annual_premium:",replace_outliers(customer_policy_details_df,'annual_premium'))
print("customer_policy_details_df with sales channel code:",replace_outliers(customer_policy_details_df,'sales channel code'))
print("customer_policy_details_df with vintage:",replace_outliers(customer_policy_details_df,'vintage'))
print("customer_policy_details_df with response:",replace_outliers(customer_policy_details_df,'response'))
print("="*50)
```

## 5. Is there any relation between Person Age and annual premium?

Hint: use correlation function (Correlation describes the relationship between two variables). 

Correlation coefficient < -0.5           - Strong negative relationship

Correlation coefficient > 0.5            -  Strong positive relationship

0.5 < Correlation coefficient < 0.5   - There is no relationship. 

```python
import numpy as np
cor_coeff=np.corrcoef(master_table['age'],master_table['annual_premium'])
print(cor_coeff)
print(cor_coeff[1][0])
#there is a no relationship  because cor_coeff is between the -0.5 and +0.5
```

```python
if cor_coeff[1][0]<-abs(0.5) and cor_coeff[0][1]<-abs(0.5):
    print("Strong negative relationship")
elif cor_coeff[1][0]>0.5 and cor_coeff[0][1]>0.5:
    print("Strong positive relationship")
else:
    print("There is no relationship")
```

```python
coo1 = master_table['age'].corr(master_table['annual_premium'])
#print('there is no relationship=>',coo1)
if coo1 < -0.5 :
    print('Strong negative relationship')
elif coo1 > 0.5:
    print('Strong positive relationship')
elif coo1 >- 0.5 and coo1< 0.5 :
    print('There is no relationship')
```
